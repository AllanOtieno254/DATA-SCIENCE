# **MATPLOTLIB**
**Line plots**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y1=[1,3,5,3,1,3,5,3,1]
y2=[2,4,6,4,2,4,6,4,2]
plt.plot(x,y1,label="SEM1")
plt.plot(x,y2,label="SEM2")
plt.plot()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("zetech")
plt.legend()
plt.show


```

**Bar plots**

```python
import matplotlib.pyplot as plt
x1=["BSE","BBIT","BCS","BED","BCM","BSCIT"]
y1=[200,300,400,250,180,180]
plt.bar(x1,y1,label="students",color='r')
plt.plot()

plt.xlabel("COURSE")
plt.ylabel("NUMBER OF STUDENTS")
plt.title("zetech university-mangu")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x1=[2,3,4,5,6]
y1=[20,40,25,10,15]

x2=[2,4,5,6,7]
y2=[10,15,10,10,30]

plt.bar(x1,y1,label="MANGU CAMOUS",color='b')
plt.bar(x2,y2,label="TRC CAMPUS",color='g')
plt.plot()

plt.xlabel("UNIT REGISTRATION")
plt.ylabel("NUMBER OF STUDENTS")
plt.title("UNIT REGISTRATION")
plt.legend()
plt.show()

```
